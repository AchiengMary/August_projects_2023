```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import random
import csv

def generate_essay(model, tokenizer, topic, max_length=200):
    """
    Generate an essay on a given topic using GPT-2.

    Args:
        model: Pretrained GPT-2 model.
        tokenizer: Corresponding tokenizer for GPT-2.
        topic (str): Essay topic.
        max_length (int): Maximum length of the essay.

    Returns:
        str: Generated essay.
    """
    prompt = f"Write an essay about: {topic}\n\n"
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(inputs, max_length=max_length, num_return_sequences=1, temperature=0.8, top_p=0.9)
    essay = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return essay

# Load GPT-2 model and tokenizer
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Example topics
topics = ["The ethics of artificial intelligence", "The impact of climate change on society", "Universal basic income"]

# Generate essays
essays = {}
for topic in topics:
    essay = generate_essay(model, tokenizer, topic)
    essays[topic] = essay

# Print essays
for topic, essay in essays.items():
    print(f"Topic: {topic}\nEssay:\n{essay}\n{'-'*50}")

import csv

# Function to save essays into a submission CSV file
def save_to_submission_file(output_file, essays):
    """
    Save generated essays into a submission CSV file.

    Args:
        output_file (str): Path to the output CSV file.
        essays (dict): Dictionary of topics and their corresponding generated essays.
    """
    with open(output_file, 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['id', 'essay'])  # Write header

        for i, (topic, essay) in enumerate(essays.items(), start=1):
            writer.writerow([i, essay])

# Output file for the submission
output_file = 'submission.csv'

# Save generated essays to the submission file
save_to_submission_file(output_file, essays)

print(f"Submission file '{output_file}' generated successfully.")
```
