```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
M = 2.52e6  # Inertia constant (J/MVA)
Xd = 0.35   # Transient reactance (p.u.)
Xl = 0.2    # Line reactance (p.u.)
Iel = 1.1   # Line current (p.u.)
P_gen = 18  # Generator output power (MW)
V_inf = 1.0 # Infinite bus voltage (p.u.)
R = 0       # Line resistance (p.u.)
dt = 0.01   # Time step (seconds)
fault_time = 6.25 / 50  # 6.25 cycles to seconds
simulation_time = 0.5   # Total simulation time (seconds)

# Initial conditions
delta_0 = 0.0    # Initial rotor angle (radians)
delta_dot_0 = 0.0  # Initial derivative of rotor angle (radians/second)
t = 0.0         # Initial time (seconds)

# Lists to store results
time_points = []
delta_values = []

# Function to calculate mechanical power input
def calculate_P_mech(t):
    if t < fault_time:
        return P_gen + 3 * (Iel ** 2) * Xl
    else:
        return P_gen

# Iterative numerical integration
while t <= simulation_time:
    # Calculate mechanical power input
    P_mech = calculate_P_mech(t)

    # Calculate derivative of rotor angle
    delta_dot = (P_mech - P_gen) / M

    # Update rotor angle
    delta_0 += delta_dot * dt

    # Update time
    t += dt

    # Store results
    time_points.append(t)
    delta_values.append(delta_0)

# Plotting the results
plt.plot(time_points, delta_values)
plt.xlabel('Time (seconds)')
plt.ylabel('Rotor Angle (radians)')
plt.title('Swing Curve with Fault Cleared at 6.25 Cycles')
plt.grid(True)
plt.show()
```
